```python
import sys, os
import numpy as np
import tkinter as tk
from tkinter import filedialog
import matplotlib.pyplot as plt
import struct
```

```python
def read_binary_file_to_numpy_array(file_path, dtype):
    try:
        with open(file_path, 'rb') as binary_file:
            data = np.fromfile(binary_file, dtype=dtype, sep='')  # Change dtype to match your data format
            return data
    except IOError as e:
        print(f"Error reading the binary file: {e}")
        return None
def check_data(data):
    data_correct = True
    for i in range(32768): #check if the data in the file is correct
        if data[i] != i:
            data_correct = False
            break
    return data_correct
# Create a tkinter GUI to select a folder
root = tk.Tk()
root.withdraw()  # Hide the main tkinter window

# Determine the path to the "data" folder within the project folder
project_folder = os.getcwd()  # Get the current working directory
folder_path = os.path.join(project_folder, 'data')

if not folder_path:
    print("No folder selected. Exiting.")
else:
    # List all files in the selected folder
    file_list = sorted(os.listdir(folder_path))
    
    
    # Exclude files with the name "TIMESTAT.bin"
    try:
        file_list = [file for file in file_list if file != "TIMESTAT.bin"]
    except:
        ValueError("Please copy the files from the uSD into ./data/ folder")
        sys.exit(1)
    # Process each file in the folder and check if the data is consistent
    for file_name in file_list:
        file_path = os.path.join(folder_path, file_name)

        # Check if the item in the folder is a file (not a subfolder)
        if os.path.isfile(file_path):
            resulting_numpy_array = read_binary_file_to_numpy_array(file_path, dtype=np.int32)

            if resulting_numpy_array is not None:
                if(check_data(resulting_numpy_array) == False):
                    print(f"Error in File: {file_name}")
    
    file_path = os.path.join(folder_path, "TIMESTAT.bin")

    # Check if the item in the folder is a file (not a subfolder)
    if os.path.isfile(file_path):
        resulting_numpy_array = read_binary_file_to_numpy_array(file_path, dtype=np.single)

        resulting_numpy_array = resulting_numpy_array[:100] #cut the data 
        if resulting_numpy_array is not None:
            # Set the figure size
            plt.figure(figsize=(15, 10))

            plt.subplot(121)
            plt.plot(resulting_numpy_array)
            plt.xlim((0,100))
            plt.ylim((0,3))
            plt.ylabel("Data Rate [MByte/s]")
            plt.xlabel("Measurement Number")
            plt.title("Data Rate Writing uSD card")

            plt.subplot(122)
            plt.boxplot(resulting_numpy_array)
            # Find the index of the lowest value in the data array
            lowest_value_index = np.argmin(resulting_numpy_array)
            plt.plot(10,np.mean(resulting_numpy_array),color ='black', label='Mean %.1f  MByte/s' % np.mean(resulting_numpy_array))
            plt.plot(10,np.median(resulting_numpy_array),color ='orange', label='Median %.1f  MByte/s' % np.median(resulting_numpy_array))

            # Highlight the lowest value with a red dot
            plt.plot(1, resulting_numpy_array[lowest_value_index], 'ro', label='Lowest Value %.1f MByte/s' % resulting_numpy_array[lowest_value_index])
            
            
            # Add a legend
            plt.legend(loc=2)
            plt.ylabel("Data Rate [MByte/s]")
            plt.ylim((0,3))
            plt.xlim((0,2))
            plt.title("Data Rate Writing uSD card Box Plot")
            plt.savefig('dataRate_Stats.png', dpi=600)


```
